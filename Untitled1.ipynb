```python
# Train and Test Path
directory_train = "Covid19-dataset/train/"
directory_test  = "Covid19-dataset/test"

# class name
categories = ["Normal","Viral Pneumonia", "Covid"]
```

```python
# func to load images
def load_random_imgs_from_folder(directory, categories):
  plt.figure(figsize=(15,15))
  for i in range(3):
    file = random.choice(os.listdir(directory))
    image_path = os.path.join(directory, file)
    img=mpimg.imread(image_path)
    ax=plt.subplot(1,3,i+1)
    ax.title.set_text(categories)
    plt.xlabel(f'Name: {file} , "Size: {img.shape}"')
    plt.imshow(img)
```

```python
for label in categories:
    load_random_imgs_from_folder(f"{directory_train}/{label}",label)
```

```python
IMG_SIZE = 150
train = []
for category in categories:
    folder = os.path.join(directory_train,category)
    label  = categories.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        img_arr  = cv2.imread(img_path)
        img_arr  = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        plt.imshow(img_arr)
        train.append([img_arr,label])
```

```python
IMG_SIZE = 150
test = []
for category in categories:
    folder = os.path.join(directory_test,category)
    label  = categories.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder,img)
        img_arr  = cv2.imread(img_path)
        img_arr  = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        plt.imshow(img_arr)
        test.append([img_arr,label])
```

```python
random.shuffle(train)
len(train) , len(test)
```

```python
x_train = []
y_train = []
for features , labels in train:
    x_train.append(features)
    y_train.append(labels)
x_test = []
y_test = []
for features , labels in test:
    x_test.append(features)
    y_test.append(labels)
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)

x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
x_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
model = Sequential()

model.add(Conv2D(16,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(32,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2)) 

model.add(Flatten())

model.add(Dense(128,input_shape = x_train.shape,activation="relu"))
model.add(Dense(3,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss = "sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(x_train,y_train,epochs=10,validation_split=0.1)
```

```python
model.summary()
```

```python
plt.plot(history.epoch,history.history["accuracy"])
plt.plot(history.epoch,history.history["val_accuracy"])
plt.legend(['train', 'test'], loc='upper left')
plt.title("Accuracy and Val_accuracy")
```

```python
plt.plot(history.epoch,history.history["loss"])
plt.plot(history.epoch,history.history["val_loss"])
plt.legend(['train', 'test'], loc='upper left')
plt.title("loss and Val_loss")
```

```python
scores = model.evaluate(x_test,y_test)
scores
```
